##### Construcción del archivo de coordenadas

```python
postal_code = list()
city = list()
address = list()
country = list()
region = list()
latitude = list()
longitude = list()

for i in df_f.cod_local.unique():
    
    print(i)
    
    id_address = df_f[df_f['cod_local']==i].Address.iloc[0]
    locator = geolocator.geocode(id_address,addressdetails=True)
    
    try:
        # Código Postal
        postal_code.append(locator.raw['address']['postcode'])
    except:
        postal_code.append("")
    try:
        # Ciudad
        city.append(locator.raw['address']['city'])
    except:
        city.append("")
    try: 
        # Direccion
        address.append(id_address)
    except:
        address.append("")
    try:
        # Pais
        country.append(locator.raw['address']['country'])
    except:
        country.append("")
    try:
        # Region
        region.append(locator.raw['address']['state'])
    except:
        region.append("")
    try:
        # Latitud
        latitude.append(locator.raw['lat'])
    except:
        latitude.append("")
    try:
        # Longitud 
        longitude.append(locator.raw['lon'])
    except:
        longitude.append("")
        
df_f = pd.DataFrame({'Cod_Local':df_f.cod_local.unique(),'Postal Code':postal_code,"City":city,"Address":address,'Country':country,'ID Country':'CL',
              'Region':region,'Latitude':latitude,'Longitude':longitude})
```
